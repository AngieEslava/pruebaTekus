```python
import pandas as pd
import os
import sys
from os import listdir
from os.path import isfile, isdir, join
from datetime import datetime, timedelta
from pandas import DataFrame
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import stats
import numpy as np
```

```python
# ".\ciudades.csv"
n = 0
rootDir = ".\data"
data_clean = pd.DataFrame()

for dirName, subdirList, fileList in os.walk(rootDir):
    n += 1
    print(n,'Directorio encontrado: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)    
        path = dirName+'\\'+fname
        df = pd.read_csv(path)
        data_clean = data_clean.append(df , ignore_index=True)
        #, ignore_index=True
#Tomado de: https://stackoverflow.com/questions/32363154/reading-multiple-csv-files-from-multiple-files-into-pandas-dataframe
```

```python
data = data_clean.copy(deep=False)
data = data.drop([ 'HardwareInteractions','StandardInteractions', 'TenantId' , 'SessionId','Id', 'DateInTicks', 'TTL'], axis=1)
data[['Fecha','Hora']] = data.Date.str.split("T",expand=True)
data = data.drop(['Hora'],axis=1)
data.Key = pd.Categorical(data.Key)
data.City = pd.Categorical(data.City)
data
```

```python
df_city =  pd.read_csv(".\ciudades.csv")
```

# Análisis Exploratorio de Datos

Para el análisis exploratorio de datos es importante tener muy en cuenta las variables y sus significados como se muestran a continuación:


    
    
    			

|Variable|Significado|
|--------|-----------|
|ArkboxInteractions|Indica la cantidad de veces que el usuario interactuó con el extraordinario panel de un solo botón|
|Duration|Describe la duración en segundos de la interacción que tuvo el usuario con la olla.|
|MovementDuration|Indica la duración en segundos del movimiento que registró la olla|
|MovementInteractions|Indica la cantidad de veces que movieron la olla|

```python
df = pd.merge(left = data, right = df_city, left_on='Key', right_on='Key')
df = df.drop(['idCity'],axis=1)
df.ArkboxInteractions = pd.to_numeric(df.ArkboxInteractions)
df.Duration = pd.to_numeric(df.Duration)
df.MovementDuration = pd.to_numeric(df.MovementDuration)
df.MovementInteractions = pd.to_numeric(df.MovementInteractions)
df.MovementInteractions.dtype
# df.Key = pd.Categorical(df.Key)
# df.City = pd.Categorical(df.City)
df
```

El análisis descriptivo de los datos muestra varias interpretaciones interesantes de las variables a analizar, por ejemplo, en la variable __ArkboxInteractions__ vemos que la media de los datos es de 167.53 veces, lo que nos diría que en promedio el usuario interactuó con el panel de control 167 veces según la medición, pero viendo la mediana en el cuartil 50 el valor es 24 veces lo que quiere decir que el 50% de los datos están por debajo de ese valor y que el valor máximo de los datos de esa variable es de 164863 interacciones el cual es bastante mayor a la media registrada, e incluso el 75% de los datos estan en un valor menor a 120 veces por lo cual no es recomendable tomar decisiones con base a los valores promedios de las variables y si proceder a analizar los outliers que pueden reflejar información valiosa sobre el comportamiento del negocio.


Realizando el mismo tipo de análisis sobre los cuartiles de los datos, para las variables como __MovementDuration__ y __MovementInteractions__ vemos que en el cuartil 50 los datos tenen el valor de cero y en el cuartil 75 de los datos se empiezan a ver valores como 5 y 1 respectivamente, saltando a un valor máximo bastante alejado de estos valores. Esto quiere decir que prácticamente el valor de 0 es lo que explicaría un comportamiento normal de los datos, reflejando también la importancia del valor cero en cada variable. Por tanto, si la mayoría de registros de la variable __MovementInteractions__ es cero quiere decir que las ollas no se mueven mucho y si los registros de __MovementDuration__ tambien son cero nos muestra que no hay duración de esos movimientos

# ¿Cuáles son los horarios entre semana y fines de semana en dónde se presentan más desplazamientos de ollas?
### ARREGLO DE FECHAS

```python
data_f = data.copy(deep=False)
data_f.Date = pd.to_datetime(data_f.Date)
data_f.Fecha = pd.to_datetime(data_f.Fecha)
data_f
```

#### Datos entre semana

```python

data_semana = data_f[data_f['Date'].apply(lambda x: x.weekday()) < 5]
data_semana.Date = data_semana['Date'].apply(lambda x: x.strftime('%H'))
data_semana.groupby('Date').sum().MovementInteractions.plot(kind='bar')
```

El horario entre semana que presenta una mayor cantidad de movimientos son las 21 horas o 9 pm, también se ve como empiezan a incrementar la cantidad de movimientos desde las dos de la tarde, lo que nos da un indicio del comportamiento de la gente con el uso de la olla con lo que se podrían plantear automatizaciones horarias para las funcionalidades de la olla en un futuro cercano si se complementa con información cualitativa sobre las necesidades de los clientes.



```python
data_semana.groupby('Date').sum().sort_values('MovementInteractions', ascending=False)
```

#### Datos Fin de semana

```python
data_finsemana = data_f[data_f['Date'].apply(lambda x: x.weekday()) >= 5]
data_finsemana.Date = data_finsemana['Date'].apply(lambda x: x.strftime('%H'))
data_finsemana.groupby('Date').sum().MovementInteractions.plot(kind='bar')
```

En el fin de semana el pico mas alto de desplazamientos se ve a las 20 horas u 8 de la noche, lo cual no es muy alejado del horario que se tenia entre semana con una distinción de una hora de diferencia en el horario de mayor uso. En este caso los otros horarios con mayor uso de la olla son a las 10 pm y a las 4 pm, muy contrario al uso entre semana donde se veia que era a las 2pm.


```python
data_finsemana.groupby('Date').sum().sort_values('MovementInteractions', ascending=False)
```
