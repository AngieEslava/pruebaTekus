```python
import pandas as pd
import os
import sys
from os import listdir
from os.path import isfile, isdir, join
from datetime import datetime, timedelta
from pandas import DataFrame
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import stats
import numpy as np
```

```python
# ".\ciudades.csv"
n = 0
rootDir = ".\data"
data_clean = pd.DataFrame()

for dirName, subdirList, fileList in os.walk(rootDir):
    n += 1
    print(n,'Directorio encontrado: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)    
        path = dirName+'\\'+fname
        df = pd.read_csv(path)
        data_clean = data_clean.append(df , ignore_index=True)
        #, ignore_index=True
#Tomado de: https://stackoverflow.com/questions/32363154/reading-multiple-csv-files-from-multiple-files-into-pandas-dataframe
```

```python
data = data_clean.copy(deep=False)
data = data.drop([ 'HardwareInteractions','StandardInteractions', 'TenantId' , 'SessionId','Id', 'DateInTicks', 'TTL'], axis=1)
data[['Fecha','Hora']] = data.Date.str.split("T",expand=True)
data = data.drop(['Hora'],axis=1)
data.Key = pd.Categorical(data.Key)
data.City = pd.Categorical(data.City)
data
```

# Cules son los horarios entre semana y fines de semana en dnde se presentan ms desplazamientos de ollas?
### ARREGLO DE FECHAS

```python
data_f = data.copy(deep=False)
data_f.Date = pd.to_datetime(data_f.Date)
data_f.Fecha = pd.to_datetime(data_f.Fecha)
data_f
```

#### Datos entre semana

```python

data_semana = data_f[data_f['Date'].apply(lambda x: x.weekday()) < 5]
data_semana.Date = data_semana['Date'].apply(lambda x: x.strftime('%H'))
data_semana.groupby('Date').sum().MovementInteractions.plot(kind='bar')
```

El horario entre semana que presenta una mayor cantidad de movimientos son las 21 horas o 9 pm, tambin se ve como empiezan a incrementar la cantidad de movimientos desde las dos de la tarde, lo que nos da un indicio del comportamiento de la gente con el uso de la olla con lo que se podran plantear automatizaciones horarias para las funcionalidades de la olla en un futuro cercano si se complementa con informacin cualitativa sobre las necesidades de los clientes.



```python
data_semana.groupby('Date').sum().sort_values('MovementInteractions', ascending=False)
```

#### Datos Fin de semana

```python
data_finsemana = data_f[data_f['Date'].apply(lambda x: x.weekday()) >= 5]
data_finsemana.Date = data_finsemana['Date'].apply(lambda x: x.strftime('%H'))
data_finsemana.groupby('Date').sum().MovementInteractions.plot(kind='bar')
```

En el fin de semana el pico mas alto de desplazamientos se ve a las 20 horas u 8 de la noche, lo cual no es muy alejado del horario que se tenia entre semana con una distincin de una hora de diferencia en el horario de mayor uso. En este caso los otros horarios con mayor uso de la olla son a las 10 pm y a las 4 pm, muy contrario al uso entre semana donde se veia que era a las 2pm.


```python
data_finsemana.groupby('Date').sum().sort_values('MovementInteractions', ascending=False)
```
